```python
# Read the file into a string
with open('Example_1_summary.txt', 'r') as file:
    file_contents = file.read()

# Print the contents
print(file_contents)
```

```python
import re

# Preprocess the text to remove timestamps (if any)
preprocessed_text = re.sub(r'\[\d+:\d+:\d+\]', '', file_contents)

# Split the text into lines
lines = preprocessed_text.split('\n')

# Print the lines
for line in lines:
    print(line)
```

```python
import nltk
nltk.download('punkt')
```

```python
# Tokenize the lines
tokenized_lines = []
all_tokens = []
for line in lines:
    tokens = nltk.word_tokenize(line)
    tokenized_lines.append(tokens)
    all_tokens.extend(tokens)

# Print the tokenized lines
for tokens in tokenized_lines:
    print(tokens)
```

```python
# Create a vocabulary (mapping of tokens to IDs)
vocab = {token: idx for idx, token in enumerate(set(all_tokens))}

# Convert tokens to IDs
id_lines = []
for tokens in tokenized_lines:
    id_line = [vocab[token] for token in tokens]
    id_lines.append(id_line)
```

```python
# Print the ID lines
for id_line in id_lines:
    print(id_line)
```

```python
# Create an inverse mapping (IDs to tokens)
inv_vocab = {idx: token for token, idx in vocab.items()}

# Decode IDs back to tokens
decoded_lines = []
for id_line in id_lines:
    tokens = [inv_vocab[id] for id in id_line]
    decoded_line = ' '.join(tokens)
    decoded_lines.append(decoded_line)
```

```python
# Print the decoded lines
for decoded_line in decoded_lines:
    print(decoded_line)
```
